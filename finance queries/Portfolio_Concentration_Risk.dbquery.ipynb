```sql
-- What percentage of our total fund is tied up in a single country ?

SELECT
  country,
  SUM(amount_invested_eur) AS country_total,
  ROUND(100.0 * SUM(amount_invested_eur) / SUM(SUM(amount_invested_eur)) OVER(),1) AS pct_of_total_fund
FROM lumiere_capital.dbt_lumiere_capital.fct_portfolio_performance
GROUP BY 1
ORDER BY pct_of_total_fund DESC;
```
