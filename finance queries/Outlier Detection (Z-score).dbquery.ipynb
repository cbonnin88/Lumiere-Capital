```sql
WITH stats AS (
  SELECT
      AVG(moic) AS avg_moic,
      STDDEV(moic) AS stddev_moic
  FROM lumiere_capital.dbt_lumiere_capital.fct_portfolio_performance
)
SELECT
  company_name,
  moic,
  (moic - stats.avg_moic) / stats.stddev_moic AS z_score
FROM lumiere_capital.dbt_lumiere_capital.fct_portfolio_performance, stats
WHERE ABS((moic - stats.avg_moic) / stats.stddev_moic) > 2
ORDER BY z_score DESC;
```
