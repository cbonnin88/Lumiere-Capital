```sql
-- Identify the 'Stars' within each green-tech category

SELECT * FROM(
  SELECT
    company_name,
    sector,
    moic,
    RANK() OVER(PARTITION BY sector ORDER BY moic DESC) AS rank_in_sector
  FROM lumiere_capital.dbt_lumiere_capital.fct_portfolio_performance
)
WHERE rank_in_sector <= 3 AND company_name IS NOT NULL;
```
