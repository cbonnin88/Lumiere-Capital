```sql
-- Total Value to Paid-In capital (TVPI)

SELECT
  SUM(current_estimated_value) AS total_value,
  SUM(amount_invested_eur) AS total_paid_in,
  (SUM(current_estimated_value) / SUM(amount_invested_eur)) AS portfolio_tvpi,
  SUM(current_estimated_value - amount_invested_eur) AS total_unrealized_profit
FROM 
  lumiere_capital.dbt_lumiere_capital.fct_portfolio_performance;
```
